```python
import os
import requests
import pandas as pd

# Create folder if it doesn't exist
os.makedirs("stakeholders", exist_ok=True)

members = soup.find_all("div", class_="mbust-member-card")

data = []

for member in members:
    img_url = member.find("img")["src"]
    name = member.find("h5").text.strip()
    position = member.find("span", class_="member-role").text.strip()

    # Create a safe filename
    filename = (
        name.replace(".", "")
            .replace(" ", "_")
            .replace("/", "_")
            + ".jpg"
    )

    image_path = os.path.join("stakeholders", filename)

    # Download image
    response = requests.get(img_url)
    if response.status_code == 200:
        with open(image_path, "wb") as f:
            f.write(response.content)

    # Save information
    data.append({
        "name": name,
        "position": position,
        "image": image_path
    })

# Save CSV
df = pd.DataFrame(data)
df.to_csv("stakeholders.csv", index=False)

print("Done!")
```

#### Download the images

```python
import os
import requests
import pandas as pd

# Create folder
os.makedirs("stakeholders", exist_ok=True)

members = soup.find_all("div", class_="mbust-member-card")

data = []

for member in members:
    # Extract information
    img_url = member.find("img")["src"]
    name = member.find("h5").text.strip()
    position = member.find("span", class_="member-role").text.strip()

    # Create a safe filename
    filename = (
        name.replace(".", "")
            .replace(" ", "_")
            .replace("/", "_")
            + ".jpg"
    )

    image_path = os.path.join("stakeholders", filename)

    # Download image
    response = requests.get(img_url)

    if response.status_code == 200:
        with open(image_path, "wb") as f:
            f.write(response.content)

        print(f"Downloaded: {filename}")
    else:
        print(f"Failed: {name}")

    # Save information
    data.append({
        "name": name,
        "position": position,
        "image": image_path
    })

# Save CSV
df = pd.DataFrame(data)
df.to_csv("stakeholders.csv", index=False)

print("Done!")
```
